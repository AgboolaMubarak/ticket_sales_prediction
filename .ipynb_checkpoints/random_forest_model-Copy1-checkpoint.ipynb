### Model training and parameter tuning

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor as GTBR
from sklearn import metrics
from sklearn.metrics import mean_absolute_error as mae
```

```python
data = pd.read_csv('train_agg1.csv')
```

```python
data.head()
```

```python
""" dropping columns that may not aid our prediction """
data = data.drop(['ride_id','month'], axis=1)

""" creating a test set from the dataset to assess the performance of the model """
test_set = data[-60*24:]

""" separating the test set from the rest of the data """
data = data[:-60*24]

""" Separate the data into features and targets and into a training and validation set"""
target_var = ['number_of_tickets']
features, targets = data.drop(target_var, axis=1), data[target_var]

train_features, train_targets = features[:-60*24], targets[:-60*24]
val_features, val_targets = features[-90*24:], targets[-90*24:]
test_features, test_targets = test_set.drop(target_var, axis=1), test_set[target_var]
```

```python
train_features.head()
```

```python
rf_model = GTBR(learning_rate=0.99, max_depth=1, n_estimators=100)
```

```python
""" training the model """
rf_model.fit(train_features,train_targets)
```

```python
""" validating performance of the model using the mean absolute error metric (mae)"""
pred = rf_model.predict(val_features)
print(mae(pred, val_targets))
```

```python
print(rf_model.score(val_features,val_targets))
```

```python
""" testing the model """
test_pred = rf_model.predict(test_features)
print(mae(test_pred, test_targets))
```

### Predicting using the test questions data

```python
df_test = pd.read_csv('test_agg1.csv')
```

```python
df_test.head()
```

```python
""" testing the model with on test set"""
test_set = df_test.drop(['ride_id', 'month'], axis=1) 
predictions = rf_model.predict(test_set)
```

```python
result = {'ride_id': df_test['ride_id'], 'number_of_tickets': np.round(predictions,0)} 
predictions_dframe = pd.DataFrame(data=result)
```
